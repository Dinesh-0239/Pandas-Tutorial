# Merging and Joining
Merging is combining two or more dataset on common column

Syntax:
pd.merge(df1,df2,on="common_col",how="type of join")

Type of join: cross,inner,outer,left,right

```python
df_customer = pd.DataFrame({
    'Cust_Id' : [1,2,3],
    'Name' : ["Dinesh","Vishal","Shreay"]
})

df_order = pd.DataFrame({
    'Cust_Id': [1,2,4],
    'Name': [1999,1499,2599]
})
print("customer",df_customer,sep="\n")
print("order",df_order,sep="\n")
```

```python
# inner join
pd.merge(df_customer,df_order,on="Cust_Id",how="inner")
```

```python
# outer join
pd.merge(df_customer,df_order,on="Cust_Id",how="outer")
```

```python
# left join
pd.merge(df_customer,df_order,on="Cust_Id",how="left")
```

```python
# right join
pd.merge(df_customer,df_order,on="Cust_Id",how="right")
```

```python
# cross join(on feild nhi hota hai)
pd.merge(df_customer,df_order,how="cross")
```

# Concatenation: row/column wise

pd.concat([df1,df2],axis=0/1,ignore_index=True)

ignore_index: used for axis=0 effective

```python
df1 = pd.DataFrame({
    "id" : [1,2,3],
    "name" : ["Dinesh","Vishal","Shreya"]
})

df2 = pd.DataFrame({
    "id" : [4,5,6],
    "name" : ["Sweta","Rushil","Sagar"]
})
df3 = pd.DataFrame({
    "likes" : ["Badminton","Basketball","Cricket"] 
})
print(df1,df2,df3,sep="\n")
```

```python
# row-wise concatenation
pd.concat([df1,df2],axis=0,ignore_index=True)
```

```python
# col-wise
pd.concat([df2,df3],axis=1)
```
