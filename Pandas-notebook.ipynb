## Creating and Saving dataframe

```python
# Creating data
data = {
    "Player" : ["Rohit","Gill","Virat","Shreyas","Rahul","Axar","Hardik","Jadeja","Kuldeep","Bumrah","Arshdeep"],
    "Score" :["79","14","142*","68","42","DNB","30*","DNB","DNB","DNB","DNB"],
    "4s" : ["6","2","12","7","4","0","2","0","0","0","0"],
    "6s" :["4","0","6","4","1","0","2","0","0","0","0"]
}

#creating data frame
df = pd.DataFrame(data)
df

#creating csv file from dataframe
df.to_csv("team-score.csv",index=False,encoding="utf8")

#creating xls file from dataframe
df.to_excel("team-score.xlsx",index=False)

#creating json file from dataframe
df.to_json("team-score.json",index=False)
```
